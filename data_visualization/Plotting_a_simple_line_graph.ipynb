```python
import matplotlib.pyplot as plt

input_values = [1, 2, 3, 4, 6]
squares = [1, 4, 9, 16, 25]

fig, ax = plt.subplots()
# ax.plot(squares, linewidth=3)
ax.plot(input_values, squares, linewidth=3)

# Set chart title and label axes.
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

# Set size of tick labels.
ax.tick_params(axis='both', labelsize=14)

plt.show()

```

```python
# scatter()

import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8')
fig, ax = plt.subplots()

# Plotting a point
ax.scatter(2, 8, s=200)

# Plotting a Series
x_values = [1, 2, 3, 4, 5]
y_values = [1, 4, 9, 16, 25]
ax.scatter(x_values, y_values, s=100)

# Set chart title and label axes.
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

# Set size of tick labels.
ax.tick_params(axis='both', which='major', labelsize=14)

plt.show()

```

```python
# Calculating Data Automatically

import random
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8')
fig, ax = plt.subplots()

x_values = range(1, 1001)
y_values = [x**2 for x in x_values]
ax.scatter(x_values, y_values, s=10)

# # using colors
ax.scatter([x+50 for x in x_values], y_values, c='red', s=30)

# # using custom colors
ax.scatter([x+100 for x in x_values], y_values, c=(0, 0.9, 0), marker="p", s=30)

# using a colormap
ax.scatter([x+150 for x in x_values], y_values, c=y_values, cmap=plt.cm.Blues, s=30)

# Set chart title and label axes.
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

# Set size of tick labels.
ax.tick_params(axis='both', which='major', labelsize=14)

ax.axis([0, 1100, 0, 1100000])


# save image 
# supported (supported formats: eps, jpeg, jpg, pdf, pgf, png, ps, raw, rgba, svg, svgz, tif, tiff, webp)
plt.savefig('squares_plot.png', bbox_inches='tight')

plt.show()  #  savefig saveshow show 
```

```python
# try 1 Cubes: A number raised to the third power is a cube. Plot the first five cubic numbers, and then plot the first 5000 cubic numbers
import matplotlib.pyplot as plt 

fig, ax = plt.subplots()

x_values = [1, 2, 3, 4, 5]
y_values = [x**3 for x in x_values]

plt.scatter(x_values, y_values, c='red')

fig, ax = plt.subplots()

x_values = range(1, 5001)
y_values = [x**3 for x in x_values]

plt.scatter(x_values, y_values, c='blue')

plt.show()


# try 2: Colored Cubes: Apply a colormap to your cubes plot.
plt.scatter(x_values, y_values, c=y_values, s=10)

plt.show()
```
